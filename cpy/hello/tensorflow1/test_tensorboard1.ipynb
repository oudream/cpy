```python
import tensorflow as tf

a = tf.constant(2.0, name='a')
b = tf.constant(3.0, name='b')
c = tf.multiply(a, b, name='c')

tf.summary.scalar('c',c)
merged = tf.summary.merge_all()

with tf.Session() as sess:
    writer = tf.summary.FileWriter("/tmp/test/2")
    writer.add_graph(sess.graph)
    summary_tf = sess.run(merged)
    writer.close()
    writer.add_summary(summary_tf)
    print("complete 2!")
```

```python
import tensorflow as tf
tf.get_variable_scope().reuse_variables()
v = tf.Variable( [2,2,3,32],name='weights')
with tf.variable_scope('variable_scope'):
    v1 = tf.get_variable('weights', [2,2,3,32])

with tf.name_scope('name_scope'):
    v2 = tf.get_variable('variable_scope/weights',[2,2,3,32])
    v4 = tf.Variable('variable_scope/weights2',[2,2,3,32])
    v5 = tf.Variable('weights2',[2,2,3,32])

with tf.name_scope('variable_scope'):
    v3 = tf.get_variable('weights')

print(tf.get_variable_scope().reuse)
print(v.name)
print(v1.name)
print(v2.name)
print(v3.name)
print(v3)
print(v4)
print(v5)

```

```python
import tensorflow as tf

k = tf.placeholder(tf.float32)

# Make a normal distribution, with a shifting mean
mean_moving_normal = tf.random_normal(shape=[1000], mean=(5*k), stddev=1)
# Record that distribution into a histogram summary
tf.summary.histogram("normal/moving_mean", mean_moving_normal)

# Setup a session and summary writer
sess = tf.Session()
writer = tf.summary.FileWriter("/tmp/histogram_example")
writer.add_graph(sess.graph)

summaries = tf.summary.merge_all()

# Setup a loop and write the summaries to disk
N = 400
for step in range(N):
    k_val = step/float(N)
    summ = sess.run(summaries, feed_dict={k: k_val})
    writer.add_summary(summ, global_step=step)
    print("complete 3 !")
```
